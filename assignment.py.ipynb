```python
import numpy as np
import pandas as pd
```

```python

df = pd.read_csv("data.csv", encoding="latin1")
df.head()


```

```python
print(df.columns)
```

```python
x = df["no2"]     # use exact column name
x = x.dropna()
```

```python
r = 102303674

ar = 0.05 * (r % 7)
br = 0.3 * ((r % 5) + 1)
print(ar)
print(br)
```

```python
z = x + ar * np.sin(br * x)
print(z)
```

```python
mu = np.mean(z)
print(mu)
sigma2 = np.var(z)
print(sigma2)

```

```python
lam = 1 / (2 * sigma2)
c = 1 / np.sqrt(2 * np.pi * sigma2)
```

```python
print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```

```python
#plotting
import matplotlib.pyplot as plt
plt.hist(z, bins=50, density=True)
plt.show()
```
